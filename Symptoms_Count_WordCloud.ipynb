```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
symptoms_df = pd.read_excel("MartySymptomsR.xls")
```

```python
symptoms_df['Symptoms'] = symptoms_df['Symptoms'].str.split(', ')
```

```python
symptoms_df['Symptoms'][0]
```

```python
symptoms_list = []
symptoms_df['Symptoms'].dropna().apply(lambda x: symptoms_list.extend(x))
```

```python
symptoms_list = [s.title().replace(" ", "") for s in symptoms_list]
```

```python
symptoms_series = pd.Series(symptoms_list)
```

```python
symptoms_series.value_counts()
```

```python
wordcloud2 = WordCloud().generate(' '.join(symptoms_series))
```

```python
plt.imshow(wordcloud2) 
plt.rcParams['figure.figsize'] = [30, 15]
plt.show()
plt.savefig('symptoms_wordcloud.png')
```
